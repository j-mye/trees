```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
import sys
sys.path.append('..')
from scripts.load_data import load_data

df = load_data()
df.head()
```

```python
import sys
sys.path.append('..')
from scripts.data_filler import fill_missing_data

df_filled = fill_missing_data(df)

# Made ONLY for visual purposes seen below
df_original = df.copy()
df_original['Height'] = df_original['Height'].fillna(0)
df_original['Crown Width'] = df_original['Crown Width'].fillna(0)
```

```python
plt.figure(figsize=(12, 10))
sns.set_style('whitegrid')

fig, axes = plt.subplots(2, 2, figsize=(12, 10), sharex='col')

sns.scatterplot(data=df_original, x='DBH', y='Height', ax=axes[0, 0], color='C0')
axes[0, 0].set_title('Height vs DBH â€” original')
axes[0, 0].set_ylabel('Height')

sns.scatterplot(data=df_original, x='DBH', y='Crown Width', ax=axes[1, 0], color='C0')
axes[1, 0].set_title('Crown Width vs DBH â€” original')
axes[1, 0].set_xlabel('DBH')
axes[1, 0].set_ylabel('Crown Width')

sns.scatterplot(data=df_filled, x='DBH', y='Height', ax=axes[0, 1], color='C1')
axes[0, 1].set_title('Height vs DBH â€” filled')

sns.scatterplot(data=df_filled, x='DBH', y='Crown Width', ax=axes[1, 1], color='C1')
axes[1, 1].set_title('Crown Width vs DBH â€” filled')
axes[1, 1].set_xlabel('DBH')

for ax in axes.flat:
    ax.label_outer()
    ax.legend(framealpha=0.6)

plt.suptitle('Original vs Filled')
plt.tight_layout()
plt.show()
```
