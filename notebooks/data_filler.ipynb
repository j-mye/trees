```python
import sys
sys.path.append('..')
from scripts.data_filler import fill_missing_data

df_filled = fill_missing_data(df)
```

```python
# Testing and visualizing the filled data
df_filled.head()

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.scatter(df['DBH'], df['Height'], label='Original')
plt.scatter(df_filled['DBH'], df_filled['Height'], label='Filled')
plt.xlabel('DBH')
plt.ylabel('Height')
plt.title('Height vs DBH')
plt.legend()
plt.subplot(1, 2, 2)
plt.scatter(df['DBH'], df['Crown Width'], label='Original')
plt.scatter(df_filled['DBH'], df_filled['Crown Width'], label='Filled')
plt.xlabel('DBH')
plt.ylabel('Crown Width')
plt.title('Crown Width vs DBH')
plt.legend()
plt.tight_layout()
plt.show()
```
